3 Schrijf een (recursieve) methode ’addFirst(value), die een nieuwe lijst retourneert met de gegeven waarde vooraan
en daarna alle elementen uit de oorspronkelijke lijst. Een aanroep van ’addFirst(5)’ op een lijst met elementen ’4’ en
’7’ bijvoorbeeld levert een nieuwe lijst op met elementen ’5’, ’4’ en ’7’. Schrijf voorbeeldcode die aannemelijk maakt
dat de methode correct werk

```python
class LinkedList:
    pass

# Empty list
class LinkedListEmpty(LinkedList):
    def __init__(self):
        pass

    # Step 2: toString() for empty list
    def toString(self):
        return ""

    # Step 3: addFirst() returns a new populated list
    def addFirst(self, value):
        return LinkedListPopulated(value, self)
    
    #step 4: remove 
    def remove(self, value):
        return self
    
    #step 5: smallest value 
    def smallest(self):
        raise Exception("Empty list has no smallest value")
    
    #step 6 sort op basis van small to large value  
    def sortSimple(self):
        return self
    



    
   


# Populated list
class LinkedListPopulated(LinkedList):
    def __init__(self, value, next):
        self.value = value
        self.next = next

    # Step 2: toString() for populated list
    def toString(self):
        return str(self.value) + " " + self.next.toString()

    # Step 3: addFirst() returns a new list with value at front
    def addFirst(self, value):
        return LinkedListPopulated(value, self)
    
    #step 4 : remove() removes 
    def remove(self, value):
        if self.value == value:
            return self.next
        else:
            return LinkedListPopulated(self.value, self.next.remove(value))
        
    #step 5 : smallest() finds the smallest value 
    def smallest(self):
        # if this is the last node
        if isinstance(self.next, LinkedListEmpty):
            return self.value
        else:
            smallest_rest = self.next.smallest()
            return min(self.value, smallest_rest)
        
    #step 6: sort into smallest to largest 
    def sortSimple(self):
        smallest_value = self.smallest()
        rest = self.remove(smallest_value)
        sorted_rest = rest.sortSimple()
        return sorted_rest.addFirst(smallest_value)





# -------------------------
# Tests for Step 2 & 3
# -------------------------

# 0 elements
empty = LinkedListEmpty()
print("Empty list:", empty.toString())   # ""

# 1 element
one = LinkedListPopulated(4, LinkedListEmpty())
print("One element:", one.toString())    # "4 "

# 2 elements
two = LinkedListPopulated(4,
      LinkedListPopulated(7, LinkedListEmpty()))
print("Two elements:", two.toString())   # "4 7 "

# Test addFirst
new_list = two.addFirst(5)
print("After addFirst(5):", new_list.toString())  # "5 4 7 "

# Original list stays unchanged
print("Original list unchanged:", two.toString()) # "4 7 "


#test for step 4
lst = LinkedListPopulated(5,
      LinkedListPopulated(4,
      LinkedListPopulated(7,
      LinkedListPopulated(4, LinkedListEmpty()))))

print("Step 4 test Original:", lst.toString())

new_lst = lst.remove(4)
print("After remove(4):", new_lst.toString())

new_lst2 = new_lst.remove(4)
print("After remove(4) again:", new_lst2.toString())

#test for step 5 
lst = LinkedListPopulated(5,
      LinkedListPopulated(4,
      LinkedListPopulated(7, LinkedListEmpty())))

print("step 5 List:", lst.toString())
print("Smallest:", lst.smallest())


#test step 6 
lst = LinkedListPopulated(5,
      LinkedListPopulated(4,
      LinkedListPopulated(7,
      LinkedListPopulated(4, LinkedListEmpty()))))

print("step 6 Original:", lst.toString())

sorted_lst = lst.sortSimple()
print("Sorted 6 :", sorted_lst.toString())



```
