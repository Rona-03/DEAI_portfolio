```python
# Alle combinaties van True/False
waarden = [True, False]

print("een\t twee\t uitkomst")

for een in waarden:
    for twee in waarden:
        uitkomst = (not een) or twee

        # Zet True/False om naar waar/onwaar
        een_txt = "waar" if een else "onwaar"
        twee_txt = "waar" if twee else "onwaar"
        uitkomst_txt = "waar" if uitkomst else "onwaar"

        print(een_txt, "\t", twee_txt, "\t", uitkomst_txt)
```
