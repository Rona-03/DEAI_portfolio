Geef een overzicht met daarin van de producten behorend tot het producttype ‘Cooking Gear’ per productnaam de totaal verkochte hoeveelheid en de gemiddelde verkoopprijs. Sorteer de uitkomst op totaal verkochte hoeveelheid. (6 kolommen, 10 rijen)

```python
# Je komt met de bovenstande info niet uit op 6 kolommen maar 4

merge16 = pd.merge(product_type1, product1, left_on = 'PRODUCT_TYPE_CODE', how = 'inner', right_on = 'PRODUCT_TYPE_CODE')
merge17 = pd.merge(order_details1, product1, left_on = 'PRODUCT_NUMBER', how = 'inner', right_on = 'PRODUCT_NUMBER')
merge18 = pd.merge(merge17, merge16, left_on = 'PRODUCT_NUMBER', how = 'inner', right_on = 'PRODUCT_NUMBER')
smerge8 = merge18[merge18['PRODUCT_TYPE_EN'] == 'Cooking Gear']

overzigt = smerge8.groupby('PRODUCT_NAME_x').agg(
    totaal_verkocht =('QUANTITY', 'sum'),
    gemiddelde_verkoopprijs =('UNIT_PRICE', 'mean'),
    product_type=('PRODUCT_TYPE_EN', 'first') #omdat je hierboven het al heb gesorteerd op Cooking Gear is alles het zelfte en gebruik je first
).reset_index()

overzigt.sort_values(by = 'totaal_verkocht', ascending = False)
```

Geef met behulp van bovenstaande hulpcode een overzicht met daarin op basis van het aantal jaar dat iemand in dienst is of een medewerker ‘kort in dienst’ (minder dan 30 jaar in dienst) of een ‘lang in dienst’ (groter dan gelijk aan 30 jaar in dienst) is. Geef daarbij per medewerker in een aparte kolom zowel ‘kort in dienst’ als ‘lang in dienst’ aan. Gebruik (wederom) een for-loop.<br>
(2 kolommen, 102 rijen) 

```python
#date.today() = geeft de datum van vandaag
#.year = haalt aleen het jaartal uit
huidig_jaar = date.today().year
df = sales_staff1.copy()
dienstverband_status = []

for datum_str in df["DATE_HIRED"]:
    # strptime() = zet de tekst om naar een datum
    date_obj = datetime.strptime(datum_str, "%Y-%m-%d")
    # berekent hoeveel jaar in dienst
    in_dienst_jaren = huidig_jaar - date_obj.year

    if in_dienst_jaren >= 30:
        status = "Lang in dienst"
    else:
        status = "Kort in dienst"
    
    dienstverband_status.append(status)

df["Dienstverband"] = dienstverband_status
resultaat4 = df[["LAST_NAME", "Dienstverband"]]
resultaat4
```
