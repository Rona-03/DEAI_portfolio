15. Hoeveel optellingen worden er uitgevoerd voordat deze methode een resultaat teruggeeft?

```python
def calculate(n):
    result = 0;
    for a in range(0, n):
        for b in range (0, n):
            for c in range (0, n):
                result += 1
    return result
```

16* Hoeveel optellingen worden er uitgevoerd voordat deze methode een resultaat teruggeeft?

```python
def calculate(n):
    result = 0;
    for a in range(0, n):
        for b in range (0, a):
            for c in range (0, b):
            result += 1
    return result
```

17* Hoeveel optellingen worden er uitgevoerd voordat deze methode een resultaat teruggeeft?

```python
def fibonacci(n):
    return 1 if n <= 1 else fibonacci(n - 2) + fibonacci(n - 1)
```
